```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns   
df = pd.read_csv(r'/Users/lawrenceliu/Documents/STAT3 Inhibitor/data_processed.csv')
days = df.Day.unique()
```

```python
#GRAPH ALL volume v. day

################## CONTROL #############
fig = plt.figure()
plt.style.use('seaborn-talk')
print (plt.style.available)

df_control = df[df["Treatment"].isin(["control#1", "control#2"])]
x = []
y = []
e = []
for day in days:
    current = df_control.loc[df_control['Day'] == day]
    current_average = current["Volume"].mean()
    current_std = current["Volume"].std()

    x.append(day)
    y.append(current_average)
    e.append(current_std)

plt.errorbar(x, y, e, label = "Control")

################## BOL4 #############
df_bol4 = df[df["Treatment"].isin(["BOL(4mg/kg)", "bol(4mg/kg)#2"])]
x = []
y = []
e = []
for day in days:
    current = df_bol4.loc[df_bol4['Day'] == day]
    current_average = current["Volume"].mean()
    current_std = current["Volume"].std()

    x.append(day+0.1)
    y.append(current_average)
    e.append(current_std)

plt.errorbar(x, y, e, label = "BOL4")

################## BOL8 #############
df_bol8 = df[df["Treatment"].isin(["BOL-8#1", "BOL-8#2"])]
x = []
y = []
e = []
for day in days:
    current = df_bol8.loc[df_bol8['Day'] == day]
    current_average = current["Volume"].mean()
    current_std = current["Volume"].std()

    x.append(day+0.2)
    y.append(current_average)
    e.append(current_std)

plt.errorbar(x, y, e, label = "BOL8")
plt.title("Treatments")
plt.xlabel("Day")
plt.ylabel("Tumor Volume")
plt.legend(loc = 'upper left')
plt.show()


```
